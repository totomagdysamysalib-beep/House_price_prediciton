```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset=pd.read_csv('housing.csv')
x=pd.concat((dataset.iloc[:,:8],dataset.iloc[:,9:]),axis=1).values
y=dataset.iloc[:,8].values
```

```python
y=y.reshape(len(y),1)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(missing_values=np.nan,strategy='mean')
imputer.fit(x_train[:,4:5])
x_train[:,4:5]=imputer.transform(x_train[:,4:5])
x_test[:,4:5]=imputer.transform(x_test[:,4:5])
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ln=ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[8])],remainder='passthrough')
x_train=np.array(ln.fit_transform(x_train))
x_test=np.array(ln.transform(x_test))
```

```python
from sklearn.preprocessing import StandardScaler
sc_x=StandardScaler()
sc_y=StandardScaler()
x_train=sc_x.fit_transform(x_train)
y_train=sc_y.fit_transform(y_train)
x_test=sc_x.transform(x_test)
y_test=sc_y.transform(y_test)
```

```python
from sklearn.svm import SVR
regressor=SVR(kernel='rbf')
regressor.fit(x_train,y_train.ravel())
```

```python
y_pred=regressor.predict(x_test)
np.set_printoptions(precision=2)
y_pred=sc_y.inverse_transform(y_pred.reshape(len(y_pred),1))
y_test=sc_y.inverse_transform(y_test.reshape(len(y_test),1))
np.concatenate(((y_pred.reshape(len(y_pred),1)),(y_test.reshape(len(y_test),1))),1)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```
