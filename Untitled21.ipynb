```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_csv('diabetes.csv')
columns_to_fix=['Glucose','BloodPressure','SkinThickness','Insulin','BMI']
dataset[columns_to_fix]=dataset[columns_to_fix].replace(0,np.nan)
```

```python

x=dataset.iloc[:,:-1].values
y=dataset.iloc[:,-1].values
```

```python
dataset.info()
```

```python
print(dataset.isnull().sum())
```

```python
dataset_corr=dataset.corr()
sns.heatmap(dataset_corr,annot=True,fmt='0.2f',linewidths=1)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=1)
```

```python
from sklearn.impute import SimpleImputer
imputer=SimpleImputer()
x_train=imputer.fit_transform(x_train)
x_test=imputer.transform(x_test)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```


 **bold text**linear Regression


```python
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression()
classifier.fit(x_train,y_train)
```

```python
y_pred=classifier.predict(x_test)
#print(np.concatenate(((y_pred.reshape(len(y_pred),1)),(y_test.reshape(len(y_test),1))),1))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

**bold text**SVM

```python
from sklearn.svm import SVC
classifier2=SVC(kernel='rbf',random_state=0)
classifier2.fit(x_train,y_train)
```

```python
y_pred=classifier2.predict(x_test)
#print(np.concatenate(((y_pred.reshape(len(y_pred),1)),(y_test.reshape(len(y_test),1))),1))
```

```python

print(classification_report(y_test,y_pred))
```

**bold text**KNN

```python
from sklearn.neighbors import KNeighborsClassifier
classifier3=KNeighborsClassifier(n_neighbors=5,metric='minkowski',p=2)
classifier3.fit(x_train,y_train)
```

```python
y_pred=classifier3.predict(x_test)
#print(np.concatenate(((y_pred.reshape(len(y_pred),1)),(y_test.reshape(len(y_test),1))),1))
```

```python

print(classification_report(y_test,y_pred))
```

**bold text**Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
classifier4=DecisionTreeClassifier(criterion='entropy',random_state=0)
classifier4.fit(x_train,y_train)
```

```python
y_pred=classifier4.predict(x_test)
#print(np.concatenate(((y_pred.reshape(len(y_pred),1)),(y_test.reshape(len(y_test),1))),1))
```

```python

print(classification_report(y_test,y_pred))
```

**bold text**Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
classifier5=RandomForestClassifier(n_estimators=10,criterion='entropy',random_state=0)
classifier5.fit(x_train,y_train)
```

```python
y_pred=classifier5.predict(x_test)
#print(np.concatenate(((y_pred.reshape(len(y_pred),1)),(y_test.reshape(len(y_test),1))),1))
```

```python

print(classification_report(y_test,y_pred))
```

**bold text**Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
classifier6=GaussianNB()
classifier6.fit(x_train,y_train)
```

```python
y_pred=classifier6.predict(x_test)
#print(np.concatenate(((y_pred.reshape(len(y_pred),1)),(y_test.reshape(len(y_test),1))),1))
```

```python

print(classification_report(y_test,y_pred))
```
